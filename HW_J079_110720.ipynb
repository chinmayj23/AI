```
import pandas as pd
import seaborn as sb
from google.colab import files
```

```
#Create a custom class DataRead to read csv file using pandas
class DataRead:
  def __init__(fname,path):
    fname.path=path
  def readfile(fname): #fn to only read the csv file
    f=pd.read_csv(fname.path)
  def RowSelect(fname): #fn to select any two rows from the csv file
    f=pd.read_csv(fname.path)
    s=f.sample(n=2)
    print(s)
```

```
class DataRead2(DataRead):
  def cor(fname):
    f=pd.read_csv(fname.path)
    plt=sb.heatmap(f.corr(),cmap='coolwarm',annot=True,cbar=True,linecolor='black')
    plt
  def RowSelect(fname): 
    f=pd.read_csv(fname.path)
    s=f.sample(n=3)
    print(s) 
```
